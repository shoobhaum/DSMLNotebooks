# **Dog Breed Identification**

This notebook builds an end to end multiclass image classifier using TensorFlow 2.0 and TensorFlow Hub.

## 1. Problem

Identifying the breed of a dog given an image of a dog

## 2. Data

We are using data from Kaggle's Dog Breed competition

https://www.kaggle.com/c/dog-breed-identification/data


## 3. Evaluation

The Evaluation is a file with prediction probabilities for each of the 120 dog breeds test images.

## 4. Features

Some information about the data:
* We're dealing with images (unstructures data) so it is best if we use deep learning/transfer learning

* There are 120 breeds of dogs (this means there are 120 classes)

* There are 10,000+ images in the training dataset (these have labels)

* There are 10,000+ images in the test dataset

### Since we've now got our training image filepaths in a list, let's prepare our labels

```python
import numpy as np
labels_list = labels['breed'].to_numpy()
labels_list
```

```python
len(labels)
```

```python
# Find unique label values
unique_breeds = np.unique(labels_list)
unique_breeds
```

```python
boolean_labels = [label == unique_breeds for label in labels_list]
```

```python
 boolean_labels[0].astype(int)
```

```python
len(boolean_labels)
```

## Creating our own validaton set

```python
# Setup X and Y variables
# We are going to start with ~1000 images and then increase as needed 

X = filenames
y = boolean_labels

# Set number of images to use for experimenting
NUM_IMAGES = 1000 #@param {type:"slider", min:1000, max:10000, step:10}
```

```python
from sklearn.model_selection import train_test_split

# Split in training and validation
X_train, X_val, y_train, y_val = train_test_split(X[:NUM_IMAGES],
                                                  y[:NUM_IMAGES],
                                                  test_size=0.2,
                                                  random_state = 42)

len(X_train), len(X_val), len(y_train), len(y_val)
```
